# Pandas Visualizations

```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing data source and establishing data frame

df = pd.read_csv(r"C:\Users\josep\OneDrive\Documents\Data Analytics\Python\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
#generate a list of potential styles to modify visualizations with (optional)

print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
#using plot to create a line graph visualization

df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
#separating out graphs using subplots

df.plot(kind = 'line', subplots = True)
```

```python
#titling the visualization

df.plot(kind = 'line', title = 'Ice Cream Ratings')
```

```python
#using x and y labels to generate axis titles

df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
#creating a bar chart

df.plot(kind = 'bar')
```

```python
#creating a stacked bar chart

df.plot(kind = 'bar', stacked = True)
```

```python
#isolating a specific column from bar chart

df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
#creating a horizontal stacked bar chart

df.plot.barh(stacked = True)
```

```python
#creating a scatter plot, where s= size of plot dots, and c = color of plot dots

df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Green')
```

```python
#creating a default histogram

df.plot.hist()
```

```python
#creating a histogram with different column sizes using bins 

df.plot.hist(bins = 20)
```

```python
#creating a box plot

df.boxplot()
```

```python
#creating a default area plot

# PLEASE NOTE: while I enjoy the clean style design fivethirtyeight provides, 
# I do acknowledge the limitations of the legend being washed out in the area charts,
# and the addition of two blue colors in the pie chart. While this would be unacceptable
# in a real-world scenario, I'm leaving as-is. I'm aware there are endless customizations
# I could make, but lack the patience to go through every style to find one that's both 
# visually appealing and works well with each viz example. My sincere apologies to purists.

df.plot.area()
```

```python
#creating an area plot with specific values to increase size for legibility

df.plot.area(figsize = (10,5))
```

```python
#creating a pie chart

df.plot.pie(y = 'Flavor Rating')
```

```python
#adjusting for legibility

df.plot.pie(y = 'Flavor Rating', figsize = (10,100))
```
